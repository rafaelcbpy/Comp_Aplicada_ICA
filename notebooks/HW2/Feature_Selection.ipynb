## Data preprocessing

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler
```

```python
# features = icols + [target_col]

# # Random spliting the training data into Train-test
# train, test = train_test_split(df[features].dropna(), test_size=0.3, random_state=42)

# Spliting the training data into Train-test by time
# It shuffles the inputs so the dataset will have an stochastic configuration
train = df.query("ano<=2017")[subset_cols].sample(frac = 1).dropna().reset_index(drop=True)
test = df.query("ano>2017")[subset_cols].dropna().reset_index(drop=True)

X_train, y_train = train.drop(columns=[target_col]), train[target_col]
X_test, y_test = test.drop(columns=[target_col]), test[target_col]
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, OneHotEncoder, StandardScaler
```

```python
ohe = make_column_transformer(
  (StandardScaler(), X_train.columns),
  # (OneHotEncoder(handle_unknown = 'ignore', drop='if_binary',sparse_output=False, dtype=float), categorical_columns),
  remainder='passthrough')

ohe.fit(X_train)
X_transformed = ohe.transform(X_train)

dummies_columns = [x.split("__")[1] for x in ohe.get_feature_names_out()]
dummies = pd.DataFrame(X_transformed, columns=dummies_columns)

X_train = dummies.astype({c: float for c in dummies_columns})

print(X_train.shape)
X_train.head()
```

## Feature Selection using Recursive feature elimination with cross-validation

```python
from xgboost import XGBRegressor
from sklearn.feature_selection import RFECV
```

```python
params = {
 'subsample': 0.9,
 'reg_lambda': 0.8,
 'reg_alpha': 1,
 'objective': 'reg:squarederror',
 'n_estimators': 300,
 'max_bin': 300,
 'learning_rate': 0.1,
 'colsample_bytree': 0.9
 }
```

```python
# Define the estimator
estimator = XGBRegressor(**params)

# Define the RFECV with importance_getter set to 'feature_importances_'
selector = RFECV(estimator, step=1, cv=5)

# Fit the selector
selector.fit(X_train, y_train)
```
